## Carga de los datos

Cargamos los ficheros de datos facilitados

## Primeros registos

```python
#hacemos primeros analisis 
#primero cuantos covids hay y cuanto del resto de enfermedades hay

sns.countplot(x = 'finding', data = data_set, palette = 'husl')
plt.xticks(rotation=80)
plt.xlabel("Diagnóstico")
plt.ylabel("Total")
plt.show()
```

```python
plt.figure(figsize=(10, 6)) 
sns.scatterplot(x='age',y='patientid',hue='finding',data=data_set)
plt.xlabel('Edad')
plt.ylabel('Número de pacientes')
plt.title('Relación entre edad y la enfermedad')
plt.show()
```

```python
#ahora estudiamos el sexo
sns.countplot(x = 'finding',hue='location', data = data_set, palette = 'husl')
plt.xticks(rotation=80)
plt.xlabel("Diagnóstico")
plt.ylabel("Total")
plt.show()
```

```python
#numero de pacientes que sobreviven al covid segun el sexo
sobreviven_covid =data_set[(data_set['finding'] == 'COVID-19') & (data_set['survival'])]
sns.countplot(x='survival',hue='sex',data=sobreviven_covid,palette='coolwarm')
plt.xlabel('Sobreviven_Covid')
plt.ylabel('Numero de pacientes')
plt.show()
```

```python
#numero de pacientes que sobreviven al covid segun la edad
portan_covid_edad= data_set[(data_set['finding'] == 'COVID-19') & (data_set['survival']=='Y')]
sns.scatterplot(x='age',y='patientid',data=portan_covid_edad)
plt.xlabel('Edad')
plt.ylabel('Número de pacientes')
plt.title('Relación entre edad y el número de pacientes que sobreviven')
plt.show()
```
